```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, 1:2]
y = df['Salary']
```

```python
Lin_reg = LinearRegression()
```

```python
Lin_reg.fit(x, y)
```

```python
poly_reg = PolynomialFeatures(degree=2)
```

```python
x_poly = poly_reg.fit_transform(x)
```

```python
x
```

```python
x_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly, y)
```

```python
%matplotlib notebook
plt.scatter(x, y)

y_prediction_linear = Lin_reg.predict(x)
y_prediction_linear2 = lin_reg2.predict(x_poly)


poly_reg3 = PolynomialFeatures(degree = 3)
x_poly3 = poly_reg3.fit_transform(x)
lin_reg3 = LinearRegression()
lin_reg3.fit(x_poly3, y)
y_prediction_linear3 = lin_reg3.predict(x_poly3)


poly_reg4 = PolynomialFeatures(degree = 4)
x_poly4 = poly_reg4.fit_transform(x)
lin_reg4 = LinearRegression()
lin_reg4.fit(x_poly4, y)
y_prediction_linear4 = lin_reg4.predict(x_poly4)

plt.plot(x, y_prediction_linear, label = 'linear regression')
plt.plot(x, y_prediction_linear2, label = 'Poly Reg. degree = 2')
plt.plot(x, y_prediction_linear3, label = 'Poly Reg. degree = 3')
plt.plot(x, y_prediction_linear4, label = 'Poly Reg. degree = 4')
plt.legend()
```
