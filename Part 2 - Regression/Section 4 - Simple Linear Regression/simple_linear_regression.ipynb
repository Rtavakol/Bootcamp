```python
import numpy as np
import sklearn
sklearn.__version__
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
x = df.iloc[:, : -1]
y = df.iloc[:, 1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 11)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
fit_line = [i * LR.coef_ + LR.intercept_ for i in x_test['YearsExperience']]
```

```python
y_predic = LR.predict(x_test)
```

```python
%matplotlib notebook
plt.scatter(x_test, y_test, label = 'test data')
plt.scatter(x_test, y_predic, label = 'y_predict')
plt.plot(x_test['YearsExperience'], fit_line, label = 'fit line')
plt.legend()
```
