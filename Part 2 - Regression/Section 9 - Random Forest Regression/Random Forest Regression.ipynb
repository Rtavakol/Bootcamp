```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
x = df.iloc[:, 1:2]
```

```python
y = df['Salary']
```

```python
regressor = RandomForestRegressor(n_estimators = 300,random_state=0)
```

```python
x_HD = np.linspace(np.min(x), np.max(x), 100)
```

```python
regressor.fit(x, y)
```

```python
y_HD_prediction = regressor.predict(x_HD)
```

```python
%matplotlib notebook
plt.scatter(x, y, label = 'Raw data')
plt.plot(x_HD, y_HD_prediction, label = 'Random Forest Model')
plt.legend()
```
