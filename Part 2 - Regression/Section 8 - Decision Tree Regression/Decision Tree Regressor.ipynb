```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
df
```

```python
x = df.iloc[:, 1:2]
y = df['Salary']
```

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x, y)
```

```python
y_prediction = regressor.predict(x)
```

```python
%matplotlib notebook
plt.scatter(x, y, label = 'Raw data')
plt.plot(x, y_prediction, label = 'Decision Tree Model')
```

```python
# In this model somethings goes wrong because we expect average over integrals, I mean we expect step like model
#Now we improve it by adding more data in our model
x_HD = np.linspace(np.min(x), np.max(x), 100) 
```

```python
y_HD_prediction = regressor.predict(x_HD)
%matplotlib notebook
plt.scatter(x, y, label = 'Raw data')
plt.plot(x_HD, y_HD_prediction, label = 'Decision Tree Model')
```
