```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.regression.linear_model as sm

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Load data
df = pd.read_csv('50_startups.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# Features
x = df.loc[:, 'R&D Spend': 'State']
x.head()
```

```python
# Targets
y = df.loc[: , 'Profit']
y.head()
```

Categorical variable encoding

```python
label_encoder = LabelEncoder()
x['State'] = label_encoder.fit_transform(x['State'])
x.head()
```

```python
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [3])], remainder='passthrough') 
x_encoded = np.array(columnTransformer.fit_transform(x), dtype = np.int)[:, 1:]
x_encoded[:3]
```

```python
#one_hot_encoder = OneHotEncoder() # categorical_features=[3]
#x_encoded = one_hot_encoder.fit_transform(x).toarray()
#x_encoded = x_encoded[:, 1:]
#x_encoded
```

```python
# Split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, train_size = 0.2, random_state = 2)
```

```python
# Linear model, fit and prediction
LR = LinearRegression()
LR.fit(x_train, y_train)
y_prediction = LR.predict(x_test)
```

```python
%matplotlib notebook
plt.figure(figsize = (6,3))
plt.bar(y_test.index, y_test, label = 'Test')
plt.bar(y_test.index, y_prediction, alpha = 0.75, label = 'Prediction')
plt.xlabel('Target Index')
plt.ylabel('Profit')
plt.legend()
plt.tight_layout()
```

```python
%matplotlib notebook
plt.figure(figsize = (6,3))
plt.bar(y_test.index, (y_test - y_prediction)/y_test)
plt.xlabel('Target Index')
plt.ylabel('Profit Difference \n from Test Values')
plt.tight_layout()
```
