# Now try a random selection

```python
import random 
N= 10000
```

```python
d = 10
ads_selected = []
```

```python
total_reward = 0
for i in range(0, N):
    ad = random.randrange(d)
    ads_selected.append(ad)
    reward = df.values[i, ad]
    total_reward = total_reward + reward
```

```python
%matplotlib notebook
plt.hist(ads_selected)
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.savefig('random_selection.png', dip = 500)
```

```python
import math
N = 10000
d = 10
adsss_selected = []
number_of_selection = [0] * d
sum_of_rewards = [0] * d
total_reward = 0
for n in range(0, N):
    ad = 0
    max_upper_bound = 0
    for i in range(0, d):
        if number_of_selection[i]>0:
            average_reward = sum_of_rewards[i]/number_of_selection[i]
            delta_i = math.sqrt(3/2 * math.log(n + 1)/ number_of_selection[i])
            upper_bound = average_reward + delta_i
        else:
            upper_bound=1e400
        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            ad = i
    adsss_selected.append(ad)
    number_of_selection[ad] = number_of_selection[ad] + 1
    reward = df.values[n, ad]
    sum_of_rewards[ad] = sum_of_rewards[ad] + reward
    total_reward = total_reward + reward
```

```python
%matplotlib notebook
plt.hist(ads_selected)
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.savefig('upper_bo_selection.png', dip = 500)
```
