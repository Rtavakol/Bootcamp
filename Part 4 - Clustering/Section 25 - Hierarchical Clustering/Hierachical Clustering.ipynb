```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
x = df.iloc[:, [3,4]].values
```

```python
x[:2]
```

```python
%matplotlib notebook
dendrogram = sch.dendrogram(sch.linkage(x, method='ward')) 
# ward method to minimize the variance inside each cluster
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.title('Dendrogram')
plt.savefig('Dendrogram plot.png', dip = 500)
```

```python
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
```

```python
y_hc = hc.fit_predict(x)
```

```python
%matplotlib notebook
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], label = 'Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], label = 'Cluster 2')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], label = 'Cluster 3')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], label = 'Cluster 4')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], label = 'Cluster 5')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.title('Clusters of Customers, Hierachical Clustering')
plt.tight_layout()
plt.savefig('Hierachical_clustering.png', dip = 500)
```
