```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
# Importing the dataset
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, 2:4]
```

```python
y = df['Purchased']
```

```python
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling
sc = StandardScaler()
x_sc_train = sc.fit_transform(x_train)
x_sc_test = sc.transform(x_test)
```

```python
# Fitting Random Forest Classification to the Training set
classifier = RandomForestClassifier(random_state = 0, n_estimators=30, criterion='entropy')
classifier.fit(x_sc_train, y_train)
```

```python
# Predicting the Test set results
y_prediction = classifier.predict(x_sc_test)
```

```python
cm = confusion_matrix(y_test, y_prediction)
```

```python
cm
```

```python
# Accuracy of the model
print(np.abs(100 * np.abs(y_prediction - y_test).sum()/y_test.count() - 100))
```

### Visualising the Training set Results

```python
%matplotlib notebook
from matplotlib.colors import ListedColormap
X_set, y_set = x_sc_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.savefig('Random Forest (Train set).png', dip = 500)
plt.show()
```

### Visualising the Test set Results

```python
%matplotlib notebook
from matplotlib.colors import ListedColormap
X_set, y_set = x_sc_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.savefig('Random Forest (Test set).png', dip = 500)
plt.show()
```

```python
%matplotlib notebook
plt.figure(figsize = (6,3))
min_samples_leafs = [1, 2, 3, 4, 5, 7]
accuracy = []
for leaf in min_samples_leafs:


    # Fitting Random Forest Classification to the Training set
    classifier = RandomForestClassifier(random_state = 0, n_estimators=30, criterion='entropy', min_samples_leaf=leaf)
    classifier.fit(x_sc_train, y_train)
    y_prediction = classifier.predict(x_sc_test)
    cm = confusion_matrix(y_test, y_prediction)
    accuracy.append(np.abs(100 * np.abs(y_prediction - y_test).sum()/y_test.count() - 100))
            
plt.plot(min_samples_leafs, accuracy, marker = 'o')
plt.xlabel('Minimum Samples Leaf')
plt.ylabel('Accuracy')
plt.semilogx()
plt.tight_layout()
```

```python
%matplotlib notebook
plt.figure(figsize = (6,3))
accuracy = []
n_estimators = np.arange(1, 100, 5)
for i in n_estimators:


    # Fitting Random Forest Classification to the Training set
    classifier = RandomForestClassifier(random_state = 0, 
                                        n_estimators=i, criterion='entropy', min_samples_leaf=2)
    classifier.fit(x_sc_train, y_train)
    y_prediction = classifier.predict(x_sc_test)
    cm = confusion_matrix(y_test, y_prediction)
    accuracy.append(np.abs(100 * np.abs(y_prediction - y_test).sum()/y_test.count() - 100))
            
plt.plot(n_estimators, accuracy, marker = 'o')
plt.xlabel('n_estimator')
plt.ylabel('Accuracy')
plt.semilogx()
plt.tight_layout()
```

```python
%matplotlib notebook
plt.figure(figsize = (6,3))
accuracy = []
m_depth = np.arange(1, 50, 1)
for depth in m_depth:


    # Fitting Random Forest Classification to the Training set
    classifier = RandomForestClassifier(random_state = 0, 
                                        n_estimators=30, criterion='entropy', min_samples_leaf=2, 
                                        max_depth = depth)
    classifier.fit(x_sc_train, y_train)
    y_prediction = classifier.predict(x_sc_test)
    cm = confusion_matrix(y_test, y_prediction)
    accuracy.append(np.abs(100 * np.abs(y_prediction - y_test).sum()/y_test.count() - 100))
            
plt.plot(m_depth, accuracy, marker = 'o')
plt.xlabel('n_estimator')
plt.ylabel('Accuracy')
#plt.semilogx()
plt.tight_layout()
```
