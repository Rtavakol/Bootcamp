```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
# Importing the dataset
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, 2:4]
```

```python
y = df['Purchased']
```

```python
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling
sc = StandardScaler()
x_sc_train = sc.fit_transform(x_train)
x_sc_test = sc.transform(x_test)
```

```python
# Fitting Naive Bayes to the Training set
classifier = GaussianNB()
classifier.fit(x_sc_train, y_train)
```

```python
# Predicting the Test set results
y_prediction = classifier.predict(x_sc_test)
```

```python
cm = confusion_matrix(y_test, y_prediction)
```

```python
cm
```

```python
# Accuracy of the model
print(np.abs(100 * np.abs(y_prediction - y_test).sum()/y_test.count() - 100))
```

### Visualising the Training set Results

```python
%matplotlib notebook
from matplotlib.colors import ListedColormap
X_set, y_set = x_sc_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.savefig('Naive Bayse (Train set).png', dip = 500)
plt.show()
```

### Visualising the Test set Results

```python
%matplotlib notebook
from matplotlib.colors import ListedColormap
X_set, y_set = x_sc_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.savefig('Naive Bayse (Test set).png', dip = 500)
plt.show()
```
